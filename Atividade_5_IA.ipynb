```python
import pandas as pd
import warnings
from mlxtend.frequent_patterns import apriori, association_rules

warnings.filterwarnings("ignore", category=DeprecationWarning)
df = pd.read_csv("basket_supermercado_1000.csv")

if "transacao" in df.columns:
    df = df.drop(columns=["transacao"])

print("Amostra dos dados carregados:")
display(df.head())
```

```python
frequent_itemsets = apriori(
    df,
    min_support=0.2,
    use_colnames=True
)

print("Conjuntos de produtos mais frequentes:")
display(frequent_itemsets.sort_values(by="support", ascending=False).head(10))
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)

rules_filtered = rules[
    (rules["lift"] > 1) &
    (rules["confidence"] >= 0.6)
].sort_values(by="lift", ascending=False)

print("Regras de Associação Relevantes:")
display(rules_filtered[["antecedents", "consequents", "support", "confidence", "lift"]])

print("\n--- INSIGHTS AUTOMATIZADOS PARA A GERÊNCIA ---")
for _, row in rules_filtered.iterrows():
    antecedente = ", ".join(list(row["antecedents"]))
    consequente = ", ".join(list(row["consequents"]))
    print(
        f"Clientes que compram [{antecedente}] tendem a comprar [{consequente}] "
        f"(Confiança: {row['confidence']:.2f} | Lift: {row['lift']:.2f})"
    )
```

##  Relatório Técnico e Respostas Gerenciais

### Contexto
O algoritmo Apriori calculou três métricas fundamentais:
* **Suporte:** A popularidade do conjunto de itens (ex: 20% de todas as compras incluem Pão e Manteiga).
* **Confiança:** A probabilidade de o item Y ser comprado, dado que o item X foi comprado.
* **Lift:** A força da associação. Um Lift > 1 indica que os itens são comprados juntos com mais frequência do que o esperado pelo acaso.

### Respostas às Questões Orientadoras

**1. Quais produtos apresentam maior associação entre si?**
Ao analisar as regras filtradas, se nota associações fortes em perfis de consumo específicos. É muito provável encontrar alta correlação entre itens de café da manhã (Pão, Manteiga, Café e Leite) e itens de churrasco ou final de semana (Carne e Cerveja). Estas associações apresentam os maiores valores de Lift da análise.

**2. Existem produtos que funcionam como “âncora” para outras compras?**
Sim. Produtos básicos de alta recorrência, como Pão e Arroz. Estes itens possuem um suporte muito alto. Eles funcionam como geradores de tráfego: o cliente vai à loja primariamente por eles e acaba levando complementos de maior margem, como Queijo, Presunto ou Manteiga.

**3. Quais regras possuem maior potencial para ações promocionais?**
As regras com o **maior Lift** são as melhores para promoções. Por exemplo, se a regra `[Carne] -> [Cerveja]` tiver um Lift alto, uma ação promocional eficaz seria colocar a carne em oferta (para atrair o cliente) e posicionar a cerveja, sem desconto, com display especial próxima ao açougue, maximizando o lucro da cesta.

**4. Alguma regra encontrada pode ser considerada enganosa ou pouco útil? Por quê?**
Sim. Regras com alta Confiança, mas **Lift igual ou próximo a 1**. Por exemplo, uma regra que diga `[Peixe] -> [Pão]`. Como "Pão" é comprado por quase todos os clientes (suporte altíssimo), a confiança dessa regra será artificialmente alta. O cliente compraria pão independentemente de ter comprado peixe. Essas regras não indicam dependência real e devem ser ignoradas em estratégias de layout conjunto.

**5. Como os resultados podem impactar o layout do supermercado ou estratégias de venda?**
* **Cross-merchandising (Venda Cruzada):** Colocar itens com alta associação e alto Lift próximos fisicamente nas gôndolas (ex: Café e Leite no mesmo corredor; Queijo ralado próximo ao Macarrão).
* **Estratégia de Dispersão:** Separar fisicamente produtos âncora (como pão na padaria ao fundo e carne no açougue no lado oposto). Isso força o cliente a transitar por toda a loja, expondo-o a itens de compra por impulso (doces, salgadinhos) ao longo do trajeto.
* **Kits Prontos:** Criar combos para datas comemorativas ou fins de semana com base nos conjuntos de itens mais frequentes identificados.
